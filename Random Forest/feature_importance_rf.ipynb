```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
df.shape
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X, y)
```
