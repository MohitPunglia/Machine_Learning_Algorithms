```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_clustering.csv")
```

```python
df.head()
```

```python
plt.scatter(df["cgpa"], df["iq"])
```

```python
from sklearn.cluster import KMeans

wscc = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit_predict(df[["cgpa", "iq"]])
    wscc.append(kmeans.inertia_)
```

```python
wscc
```

```python
plt.plot(range(1, 11), wscc)
```

```python
X = df.iloc[:, :].values
kmeans = KMeans(n_clusters=4)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
X[y_kmeans == 3, 1]
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], color="red")
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], color="blue")
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], color="green")
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], color="yellow")
```
